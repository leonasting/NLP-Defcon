## Data Cleaning

```python
dict_content_count = data_url["content"].value_counts().to_dict()
```

```python
dict_duplicate={key:val  for key, val in dict_content_count.items() if val>1}
```

```python
print("Unique Data count",len(dict_content_count) ,"Duplicate content count",len(dict_duplicate))
```

```python
ls_duplicate=[]
for index,row in data_url.iterrows():
    if row['content'] in dict_duplicate:
        ls_duplicate.append(index)
    
```

```python
print("Duplicated content count",len(ls_duplicate))
```

```python
ls_unique_index = list(filter(lambda x: x not in ls_duplicate,data_url.index))
```

Based on duplicates there is pattern which is content starting with "*"  is used for reference information

```python
ls_index_ref=data_unique.apply(lambda row: row["content"].strip()[0]=="*",axis=1)
```

```python
print("Filtered index based on first character as asterix",sum(ls_index_ref))
```

```python
data_unique.loc[list(ls_index_ref),:]
```

```python
data_unique=data_unique.loc[list(not x for x in ls_index_ref),:]
```
