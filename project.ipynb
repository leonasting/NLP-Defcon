```python
import requests
from bs4 import BeautifulSoup
page_num = 3
url3="https://www.defense.gov/News/Contracts/?Page="+str(page_num)      
page = requests.get(url3)
```

```python
if page.status_code == 200:
  sp = BeautifulSoup(page.text, 'html.parser')
else:
  print("Unable to retrieve data")
```

```python
anchors = sp.findAll("listing-titles-only")
urls = [[element["article-url"],element["publish-date-ap"]] for element in anchors]
```

```python
for url, date_contract in urls:
  try:
      contract_page = requests.get(url)
      if contract_page.status_code == 200:
        sp2 = BeautifulSoup(contract_page.text, 'html.parser')
        div_content=sp2.find("div", {"class":"body"})
        paragraphs = div_content.findAll("p")
        page = page_num
        cnt = 1
        category = ""
        content = []
        for p in paragraphs:
          strong_element = p.find("strong")
          if strong_element:
            category=strong_element.text
          else:
            content.append({"para_id":"pg_"+str(page)+"_para_"+str(cnt),
                        "content":p.text,
                        "category":category,
                        "publish_date_ap":date_contract
                          })
            cnt += 1
  except: pass
```
